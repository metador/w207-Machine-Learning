```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GMM
from matplotlib.colors import LogNorm

from matplotlib.colors import ListedColormap
import matplotlib.cm as cm
import warnings
warnings.filterwarnings('ignore')
```

Load feature names.

```python
feature_names = []
with open('./mushroom.map') as fmap:
    for line in fmap:
        #print line
        [index, name, junk] = line.split()
      
        feature_names.append(name)

print 'Loaded feature names:', len(feature_names)
print feature_names
```

Load data. The data is sparse in the input file, but there aren't too many features, so we'll use a dense representation, which is supported by all sklearn objects.

```python
X, Y = [], []

with open('mushroom.data') as fdata:
    for line in fdata:
        
        items = line.split()
        
        Y.append(int(items.pop(0)))
        x = np.zeros(len(feature_names))
        
        for item in items:
            feature = int(item.split(':')[0])
            x[feature] = 1
        X.append(x)

# Convert these lists to numpy arrays.
X = np.array(X)
Y = np.array(Y)

# Split into train and test data.
train_data, train_labels = X[:7000], Y[:7000]
test_data, test_labels = X[7000:], Y[7000:]

# Check that the shapes look right.
print train_data.shape, test_data.shape
#print train_data
```

(1) Run Principal Components Analysis on the data. Show what fraction of the total variance in the training data is explained by the first k principal components, for k in [1, 50].

```python
def P1():
### STUDENT START ###
    k_list = range(1,51)
    for k in k_list:
        #Converting to PCA components
        pc = PCA(n_components=k)
        pc.fit(train_data)
        print "The faction of total variance explained with k = %d is %3.6f" %(k, sum(pc.explained_variance_ratio_))

### STUDENT END ###

P1()
```

(2) PCA can be very useful for visualizing data. Project the training data down to 2 dimensions and plot it. Show the positive (poisonous) cases in blue and the negative (non-poisonous) in red. Here's a reference for plotting: http://matplotlib.org/users/pyplot_tutorial.html

```python
def P2():
### STUDENT START ###
    k=2
    pc = PCA(n_components=k)
    train_two_comps = pc.fit_transform(train_data)
    ## Ploting the graph
    plt.scatter(pc.components_[0],pc.components_[1])
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    
    plt.scatter(train_two_comps[:,0],train_two_comps[:,1],c=train_labels, cmap=cm_bright)

### STUDENT END ###

P2()
```

(3) Run KMeans with [1,16] clusters over the 2d projected data. Mark each centroid cluster and plot a circle that goes through the most distant point assigned to each cluster.

```python
def P3(Xpair,N):
### STUDENT START ###
    k=2
    # Converting data to components
    pc = PCA(n_components=k)
    train_two_comps = pc.fit_transform(train_data)
    Xpair = train_two_comps
    
    #applying k-means
    km = KMeans (n_clusters=N, init='k-means++')
    clstrs = km.fit (Xpair)
    
    myColors = ["green","blue","magenta","cyan","red","gold","0.2","0.4","pink","brown","darkred","darkblue","darkblue","maroon","grey","olive",]

    myClCols = []
    ##Ploting the clusters 
    for ii in np.arange (0, len(clstrs.labels_)):
        color_index = clstrs.labels_[ii]
        myClCols.append(myColors[color_index])
    fig = plt.figure(figsize = (5, 5))
    ax = fig.add_subplot(1, 1, 1)
    plt.scatter (Xpair[:,0], Xpair[:,1], c = myClCols , facecolors='none')
    
    radius = np.zeros(N)
    ##Getting the radius of the furtherest point for a given cluster
    for i in range(len(clstrs.labels_)):
        cls_no = clstrs.labels_[i]
        distance = ((Xpair[i,0]- clstrs.cluster_centers_[cls_no,0])**2 + (Xpair[i,1]- clstrs.cluster_centers_[cls_no,1])**2)**0.5
        if distance > radius[cls_no]:
            radius[cls_no] = distance
    #plotting thier center points
    plt.scatter (clstrs.cluster_centers_[:,0], clstrs.cluster_centers_[:,1], c = 'orange',marker='+',s=100, linewidths=3)
    #ploting their cluster
    for i in range(len(clstrs.cluster_centers_)):
        ax.add_artist(plt.Circle ((clstrs.cluster_centers_[i,0], clstrs.cluster_centers_[i,1]), radius[i], color = 'black',fill=False))

    plt.title (str(N) + " clusters")
    plt.show()

### STUDENT END ###
for nn in range (1, 16):
    P3(train_data, N = nn)
```

(4) Fit a Gaussian Mixture Model for the positive examples in your 2d projected data. Plot the estimated density contours as shown here: http://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_pdf.html#example-mixture-plot-gmm-pdf-py. Vary the number of mixture components from 1-4 and the covariance matrix type ('spherical', 'diag', 'tied', 'full').

```python
def P4():
    
### STUDENT START ###
    k_list = range(1,5)
    #converting the data into points components
    
    pc = PCA(n_components=2)
    train_two_comps = pc.fit_transform(train_data)
    #taking only positive data
    X_train = train_two_comps[train_labels==1]
    i=0
    for k in k_list:

        covariance_list = ['spherical', 'tied', 'diag', 'full']
        for covariance_type in covariance_list:
            i=i+1
            clf = GMM(n_components=k, covariance_type=covariance_type)
            clf.fit(X_train)

            # display predicted scores by the model as a contour plot
            
            x = np.linspace(-4.0, 4.0)
            y = np.linspace(-3.0, 3.0)
            X, Y = np.meshgrid(x, y)
            XX = np.array([X.ravel(), Y.ravel()]).T
            Z = -clf.score_samples(XX)[0]
            Z = Z.reshape(X.shape)
            plt.figure(figsize=(7,5))
            #plt.subplot(len(covariance_list),len(k_list),i)
            CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0),
                             levels=np.logspace(0, 2, 20))
            CB = plt.colorbar(CS, shrink=0.8, extend='both')
            plt.scatter(X_train[:, 0], X_train[:, 1], .8)

            plt.title('For k = %d s = %s'%(k,covariance_type))
            plt.axis('tight')
            plt.show()
### STUDENT END ###

P4()
```

(5) Fit two 4-component full covariance GMMs, one for the positive examples and one for the negative examples in your 2d projected data. Predict the test examples by choosing the label for which the model gives a larger probability (use GMM.score). What is the accuracy?

```python
def P5():

### STUDENT START ###

    pc = PCA(n_components=2)
    
    train_two_comps = pc.fit_transform(train_data)
    ##Separating postive and negative samples
    X_pos = train_two_comps[train_labels==1]
    X_neg = train_two_comps[train_labels==0]
    
    #Creating components for test data
    test_two_comps = pc.transform(test_data)
    
    clf_pos = GMM(n_components=4, covariance_type='full').fit(X_pos)
    score_pos = clf_pos.score(test_two_comps)
    clf_neg = GMM(n_components=4, covariance_type='full').fit(X_neg)
    
    score_neg = clf_neg.score(test_two_comps)
    
    #Comparing the scores
    predict = score_pos >score_neg
    
    print "The accuray is given by"
    print 1.00*sum(predict==test_labels)/len(test_labels)
    
### STUDENT END ###

P5()
```
